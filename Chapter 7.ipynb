```python
ch7 = imread('images/ch7.jpg')
plt.figure(figsize=(20, 5))
plt.axis('off')
plt.title('The street where he lives')
plt.tight_layout()
plt.imshow(ch7);
```

I see a carpark here, and I think it matches the one I saw on a map of the city. I'll try to detect the corners I want to match here in this source image.

```python
plt.figure(figsize=(20, 12))
plt.imshow(ch7)
plt.plot(460,960, 'or')
plt.plot(15,1145, 'or')
plt.plot(335,1265, 'or')
plt.plot(645,995, 'or');
```

```python
src = np.array([460, 960,
                15, 1145,
                335, 1265,
                645, 995,
]).reshape((4, 2))
```

This is the map. I try to get the corners of this carpark as well. I have a hunch that they are one and the same as in the image above.

```python
carpark = imread('images/carpark.png')
plt.figure(figsize=(20, 12))
plt.imshow(carpark)
plt.plot(195,470, '.r')
plt.plot(160,655, '.r')
plt.plot(300,670, '.r')
plt.plot(315,485, '.r');
```

```python
dst = np.array([195, 470,
                160, 665,
                300, 670,
                315, 485,
]).reshape((4, 2))*2
```

Using homography transformation, we change our perspectives a little bit. What I'm doing is something that isn't hurting anyone, really. Aside from the time I collected blood. But that was one time! I needed to do that. And I won't hurt him again, I promise. This projective transformation just confirms my guess.

```python
tform = transform.estimate_transform('projective', src, dst)
tf_img = transform.warp(ch7, tform.inverse)
fig, ax = plt.subplots(figsize=(20, 12))
ax.axis('off')
ax.imshow(tf_img)
_ = ax.set_title('Projective Transformation')
```
